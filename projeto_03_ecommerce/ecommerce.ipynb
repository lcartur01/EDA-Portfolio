## Importing libreries to be used
The librearies to be used will be Pandas and Matplotlib

```python
# Manipulation libraries
import pandas as pd

# Data visualization library
import matplotlib.pyplot as plt
```

## Load the Dataset
In this step, we will load the dataset, which is in CSV format, and perform a preliminary analysis of the data.

```python
# Loading the dataset
df = pd.read_csv('amazon_sales_dataset.csv', sep=None, engine='python')

# Show basic information
print("✅ The dataset has been loaded")
print(f"\n1️⃣ Total number of lines:....{len(df):,}")
print(f"\n2️⃣ Total number of columns:..{df.shape[1]}")
print(f"\n3️⃣ Memory used:..............{df.memory_usage(deep=True).sum()/1024:.2f} KB")
```

```python
# Random sample of the data

print("🎲 Random sample of the data:")
df.sample(20)
```

```python
# Showing null data

print("❎ Null data per column:")
df.isnull().sum()
```

```python
# Showing data type

print("🧩 Data types by column:")
df.dtypes
```

```python
# Showing duplicate data

print("🖨 Lines with duplicate data:")
df.duplicated().sum()
```

```python
# Showing basic dataset information

print("🔎 Basic information about the columns relating to values:")
df[["unit_price","discount","shipping_cost","total_sales"]].describe()
```

```python
# Transforming date values ​​into a manageable format

df['order_date1'] = pd.to_datetime(df['order_date'])
df['ship_date1'] = pd.to_datetime(df['ship_date'])
df['delivery_date1'] = pd.to_datetime(df['delivery_date'])
print("📶 Dates converted successfully!")
```

## Analysis of all requests.
In this step, we will analyze all the information. Let's go!

```python
# Showing the total gross sales

print("💎 What was the total gross sales value during the period?")
gross_total = (df['quantity']*df['unit_price']).sum()
gross_total
```

```python
# Showing the total sales value

print("💲 What was the total net sales value during the period?")
net_total = df['total_sales'].sum()
net_total
```

```python
# Showing the total value of the discounts

print("💸 What is the total value of the discounts applied during the period?")
discounts_applied = ((df['quantity']*df['unit_price'])*df['discount']).sum()
discounts_applied
```

```python
# Showing the percentage of shipping costs relative to the total sales value

print("🚢 What percentage of total sales does shipping cost represent?")
df['shipping_cost'].sum()/(gross_total-discounts_applied)*100
```

## Analysis of the characteristics of the requests
In this step, we will analyze some characteristics of the orders from this period!

```python
# Showing customers who have purchased more than once

print("🛍 How many customers made more than one purchase during the period?")
(df['customer_id'].value_counts()>1).sum()
```

```python
# Showing the categories of best-selling items, from lowest to highest

print("🛒 What were the store's best-selling items during that period?")
df.groupby('category')['order_id'].count().sort_values()
```

```python
# Showing the most commonly used payment methods, from smallest to largest

print("💳 What were the most commonly used payment methods during that period?")
df.groupby('payment_method')['order_id'].count().sort_values()
```

## Analysis of order values
In this step, we will analyze the total sales!

```python
# Showing the average ticket price for the period

print("🎫 What was the average sales ticket for the period?")
df['total_sales'].mean()
```

```python
# Showing sales by day

print("📆 What is the daily sales value?")
df.groupby('order_date')['total_sales'].sum()
```

```python
# Showing total sales per month
print("💹 What is the monthly sales volume?")
df['month'] = df['order_date1'].dt.month
df.groupby('month')['total_sales'].sum()
```

## Final analysis of orders by product
In this step, we will analyze the products based on the number of orders and their revenue!

```python
# # Showing the best-selling products based on the number of orders, from lowest to highest

print("🏅 What types of products are most requested during this period?")
df.groupby('product_name')['order_id'].count().sort_values()
```

```python
# Showing the best-selling products by price, from lowest to highest

print("📈 What types of products generated the most revenue during the period?")
df.groupby('product_name')['total_sales'].sum().sort_values()
```

## Graphical analysis of the table
A brief analysis with graphs from the table!

```python
# Let's show a line graph of the sales days!

daily_sales = df.groupby('order_date1')['total_sales'].sum()

plt.figure(figsize=(7,3))
plt.plot(daily_sales.index, daily_sales.values)
plt.title('Sales per day')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Let's show a bar chart of the product subcategories!

categories_sold = df.groupby('sub_category')['order_id'].count()

plt.figure()
plt.bar(categories_sold.index, categories_sold.values)
plt.title('Sales by subcategory')
plt.xticks(rotation=90)
plt.show()
```

```python
# Let's show a distribution chart of the average ticket price.

plt.figure()
plt.hist(df['total_sales'], bins=30)
plt.title('Average ticket distribution')
plt.show()
```
